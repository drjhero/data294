Now let's cover some key functions that will come up often when wrangling data.

```python
# Applying functions to elements in panda objects
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
frame
```

```python
np.abs(frame) # gets the absolute value of each element in frame
```

```python
# We can apply custom functions too
def f1(x):
  return x.max() - x.min() # a per column op basically
frame.apply(f1)
```

```python
# We can apply the operation per row too
frame.apply(f1, axis="columns") # i.e., apply across the columns
```

```python
# The custom functions can also return Series
def f2(x):
  return pd.Series([x.min(), x.max()], index=["min", "max"])
frame.apply(f2)
```
